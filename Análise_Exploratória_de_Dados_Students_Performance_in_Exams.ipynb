## 1. Importando bibliotecas

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Configurações de estilo

```python
theme_color = 'pastel'
sns.set(style='whitegrid', palette=theme_color)
```

## 2. Carregando os dados

```python
url = ('/content/drive/MyDrive/1 - Treinos Ciência de Dados 2025/Análise Exploratória de Dados - Students Performance in Exams/StudentsPerformance.csv')
df = pd.read_csv(url)
```

## 3. Visão geral do dataset

```python
print('\nFormato do DataFrame:', df.shape)
df.head()
```

## 4. Estatísticas descritivas

```python
df.describe()
```

## 5. Verificando dados faltantes

```python
print("\nVlores ausentes: ")
print(df.isnull().sum())
```

## 6. Distribuição das notas

```python
fig, axs = plt.subplots(1, 3, figsize=(18, 5))
sns.histplot(df['math score'], kde=True, ax=axs[0], color='skyblue')
sns.histplot(df['reading score'], kde=True, ax=axs[1], color='lightgreen')
sns.histplot(df['writing score'], kde=True, ax=axs[2], color='salmon')
axs[0].set_title('Distribuição de notas de Matemática')
axs[1].set_title('Distribuição de notas de Leitura')
axs[2].set_title('Distribuição de notas de Escrita')
plt.tight_layout()
plt.show()
```

## 7. Boxplots por gênero

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='gender', y='math score', data=df)
plt.title('Notas de Matemática por Gênero')
plt.show()
```

## 8. Efeito do preparo do curso

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='test preparation course', y='math score', data=df)
plt.title('Efeito do curso preparatório nas notas de Matemática')
plt.show()
```

## 9. Média geral por grupo étnico

```python
group_avg = df.groupby('race/ethnicity')[['math score','reading score','writing score']].mean()
group_avg.plot(kind='bar', figsize=(10,6))
plt.title('Médias por grupo étnico')
plt.ylabel('Nota média')
plt.xticks(rotation=0)
plt.show()
```

## 10. Correlação entre as notas

```python
print("""
Principais insights:
- Estudantes que fizeram curso preparatório têm, em média, notas mais altas.
- Há uma correlação positiva entre as três áreas avaliadas.
- Diferenças pequenas mas notáveis entre grupos étnicos e gênero.

Possíveis próximos passos:
- Aplicar clustering para segmentação de perfis.
- Usar regressão para prever nota final com base nas características.
""")
```
